### Data Cleaning and Manipulation

```python
data.duplicated().sum()
```

```python
# Droping the unnecessary columns
data.drop(columns={'Unnamed: 0.1'},inplace=True)
```

```python
# Renameing the column
data.rename(columns={'Unnamed: 0' : 'Id'},inplace=True)
```

```python
data['Platform'] = data['Platform'].str.replace("whatsapp", "whatsapp",0)
```

```python
data['Platform'].value_counts()
```

### Data Analysis and Visualization

```python
data['Platform'].value_counts(normalize=True).plot.pie(title= 'Platform wise Distribution',autopct='%1.1f%%',figsize=(8, 5))
plt.show()
```

```python
sns.kdeplot(data=data, x='Likes', hue='Platform', fill=True)
plt.title('Platform wise Likes Distribution')
plt.show()
```

```python
sns.kdeplot(data=data, x='Retweets',hue='Platform',fill=True)
plt.title('Platform wise Retweets Distribution')
plt.show()
```

```python
sns.scatterplot(x='Likes', y='Retweets', data=data,hue='Platform')
plt.title('Likes v/s Retweets')
plt.show()
```

```python
data['Country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Tweeting Countries")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=data, x = 'Year',hue='Platform')
plt.title("Tweet Count Per Year")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=data, x = 'Month',hue='Platform')
plt.title("Monthly Trends")
plt.show()
```
